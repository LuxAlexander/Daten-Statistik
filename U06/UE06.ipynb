### Wie könnte sich die Anzahl gewonnener Medaillen Österreichs bei den nächsten Winterspielen auf Basis bisheriger Ergebnisse entwickeln?

```python
# Filtern auf österreischische Athlet:innen
df_athletes_austrian = df_athletes[df_athletes['NOC'] == 'AUT']
```

```python
df_athletes_austrian
```

```python
# Filtern auf Winterspiele
df_athletes_austrian_in_winter = df_athletes_austrian[df_athletes_austrian['Season'] == 'Winter']
```

```python
df_athletes_austrian_in_winter
```

```python
# Medaillengewinne pro Jahr im gefilterten DataFrame berechnen
value_counts = df_athletes_austrian_in_winter.dropna(subset=['Medal'])['Year'].value_counts()
```

```python
value_counts
```

```python
# Medaillengewinne nach Jahr sortieren (wichtig für die Berechnung der Regression!)
value_counts = value_counts.sort_index()
```

```python
# Zuweisung von x, y, und n für Scatterplot & Berechnung der Regression
x = value_counts.index
y = value_counts.values
n = len(x)
```

```python
# Scatterplot Österreichs gewonnener Medaillen vs. Austragungsjahr der olympischen Winterspiele
fig = plt.figure(
    dpi=250
)
plt.scatter(x, y)
plt.ylabel('Gewonnene Medaillen')
plt.xlabel('Austragungsjahr')
plt.xticks(np.arange(x.min(), x.max()+4, 4), rotation=90)
plt.title('Gewonnene Medaillen Österreichs bei olympischen Winterspielen in athlete_events.csv')
plt.show()
```

```python
# Berechnung von Regressionen 1. bis 12. Ordnung aus UE03 auf die Daten

fig = plt.figure()
gs = fig.add_gridspec(3, 4, hspace=0, wspace=0)
axes = gs.subplots(sharex='col', sharey='row')
fig.set_figheight(12)
fig.set_figwidth(16)
fig.set_dpi(150)
fig.suptitle('Regressionskurven 1. bis 12. Ordnung')
fig.supxlabel('X')
fig.supylabel('Y')

# calling 'flat' makes iterating over subplots easier for us.
for idx, ax in enumerate(axes.flat):
    # get order for Polynomial as k
    k = idx+1
    # plot x vs y
    ax.scatter(x,y)
    
    # fit Polynomial of k-th Order with full=True to return sum of squared residuals
    poly, [rss, _, _, _] =  Polynomial.fit(x,y, deg=k, full=True)
    
    # unpack sum of squared residuals from array
    rss = rss[0]
    
    # plot regression line
    ax.plot(x, poly(x), color='r', label=f'{k}. Ordnung')
    
    # show legend
    ax.legend(loc='lower right')
    
    # first we calculate s_hat_squared as specified in the formula:
    s_hat_sq = 1/(n-k) * sum((y - poly(x))**2)
    

    # then we calculate sy_squared, which is actually just the variance of y with delta degrees of freedom set to 1, as specified in the formula:
    # parameter ddof=1 ensures that sum is divided by n - 1 instead of n
    sy_sq = np.var(y, ddof=1)
    # sy_sq = 1/(n-1) * sum((y - np.mean(y))**2) does the same
    
    # lastly, we calculate r_bar_xy^2:
    r_bar_xy_sq = 1 - (s_hat_sq / sy_sq)
    
    # show rss and r_xy^2 in upper right corner
    ax.text(x.min(), y.max()-1, f'RSS: {rss:.3f}')
    ax.text(x.min(), y.max()-3, f'r_xy^2: {r_bar_xy_sq:.3f}')
    
    # some more styling to make plot fancier
    ax.label_outer()
    ax.tick_params(axis='x', labelrotation = 90)
        


fig.tight_layout()
plt.show()
```

```python
# Auswahl von k nach den Bekannten Kriterien (r_xy^2 möglichst groß bei minimalem k, kleine RSS)
k = 4
```

```python
# Fit der Regression mit gewähltem k
poly, [rss, _, _, _] =  Polynomial.fit(x,y, deg=k, full=True)
```

```python
# Definition des Wertebereichs für die Regression inkl. Vorhersage bis 2036
x_poly = np.append(x, np.arange(2014, 2040, 4))
y_poly = poly(x_poly)
```

```python
# Forecasting mittels Regression 4. Ordnung auf Basis von Ergebnissen der letzten Winterspiele
fig = plt.figure(
    dpi=250
)
plt.scatter(x, y, label='Historische Daten')
plt.ylabel('Gewonnene Medaillen')
plt.xlabel('Austragungsjahr')
plt.plot(x_poly, y_poly, color='r', label=f'Lin. Reg {k}. Ordnung')
plt.xticks(np.arange(x_poly.min(), x_poly.max()+4, 4), rotation=90)
plt.yticks(np.arange(0, y_poly.max()+5, 5))
plt.title('Forecast für gewonnene Medaillen Österreichs bei olympischen Winterspielen auf Basis von athlete_events.csv')
plt.grid(axis='y')
plt.legend(loc='lower right')
plt.tight_layout()
plt.show()
```

### Wie hoch ist die Wahrscheinlichkeit eine Medaille im Eishockey zu gewinnen (inkl. 95% Konfidenzintervall) unter Spieler:innen des Eishockeyturniers?

```python
# Filtern des Sports auf die Kategorie Ice Hockey
df_athletes_icehockey = df_athletes[df_athletes['Sport'] == 'Ice Hockey']
```

```python
df_athletes_icehockey
```

```python
# Anzahl aller Teilnehmer:innen am Eishockeyturnier über die Jahre kann bspw. dem Attribut .shape entnommen werden.
df_athletes_icehockey.shape
```

```python
# Definieren von n über DataFrame.shape
n = df_athletes_icehockey.shape[0]
```

```python
# Filtern aller Medaillengewinner:innen über die Bedingung, dass in ihrer jeweiligen Zeile für die Spalte Medal ein Wert != NaN gesetzt wurde
df_athletes_icehockey[~df_athletes_icehockey['Medal'].isna()]
```

```python
# Anzahl aller Medaillengewinner:innen beim Eishockeyturnier über Zugriff auf .shape
medallists = df_athletes_icehockey[~df_athletes_icehockey['Medal'].isna()].shape[0]
```

```python
# Schätung von p_dach:
p_hat = medallists/n
```

```python
# Berechnung des Konifdenzintervalls
alpha = 0.05
k = stats.norm.ppf(1-alpha/2)
l=k/(2*np.sqrt(n))
p_minus = p_hat-l
p_plus = p_hat+l

print(f'Geschätzte Medaillenchance einer Eishockeyspieler:in bei den olympischen Winterspielen (n={n}):\t{p_hat*100:.1f}%')
print(f'Wahre Medaillenchance einer Eishockeyspieler:in bei den olympischen Winterspielen liegt mit {(1-alpha)*100:.0f}%iger Sicherheit zwischen {p_minus*100:.1f}% und {p_plus*100:.1f}%')
```

### Ändert sich die Wahrscheinlichkeit eine Medaille im Eishockey zu gewinnen (inkl. 95% Konfidenzintervall) für Spieler:innen des Eishockeyturniers signifikant über die Jahre?

```python
# Filtern der Spalte Year auf einzigartige Einträge per unique(...)-Funktion sowie aufsteigende Sortierung dieser Einträge über np.sort(...)
years = np.sort(df_athletes_icehockey['Year'].unique())
```

```python
# Visualisierung der Entwicklung der Medaillenchancen auf Basis von UE05.c)
alpha = 0.05
k = stats.norm.ppf(1-alpha/2)

fig = plt.figure()
fig.set_dpi(150)
fig.set_figheight(8)
fig.set_figwidth(12)
for year in years:
    df_athletes_yearly = df_athletes_icehockey[df_athletes_icehockey['Year'] == year]
    n = df_athletes_yearly.shape[0]
    p_hat = df_athletes_yearly[~df_athletes_yearly['Medal'].isna()].shape[0] / n
    l=k/(2*np.sqrt(n))
    interval = (p_hat-l, p_hat+l)
    plt.plot((year, year), interval, 'ro-', alpha=0.5)
    plt.plot(year, p_hat, 'bo')

plt.xticks(years);
plt.grid()
plt.title('Entwicklung der Medaillenchancen einer Eishockeyspieler:in bei den olympischen Winterspielen lt. athlete_events.csv')
plt.xlabel('Jahre')
plt.xticks(rotation=90)
plt.ylabel('Geschätzte Medaillenchancen [%] inkl. 95% Konfidenzintervall')
plt.tight_layout()
plt.show()
```
